```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
tips['total_bill'].plot.hist()
```

```python
tips[['total_bill', 'tip']].plot.hist(bins = 20, alpha = 0.5)
```

```python
tips['tip'].plot.kde()
```

```python
tips.plot.scatter(x = 'total_bill', y = 'tip')
```

```python
tips.plot.hexbin(x = 'total_bill', y = 'tip')
```

```python
tips.plot.hexbin(x = 'total_bill', y = 'tip', gridsize = 10)
```

```python
tips.plot.box()    # 숫자로 이루어진 셀들을 자동으로 표현
```

```python
tips.info()    # float, int...
```

```python
tips[['total_bill', 'tip']].plot.box()
```
