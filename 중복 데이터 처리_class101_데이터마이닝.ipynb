```python
import pandas as pd

billboard = pd.read_csv('billboard.csv')
billboard
```

```python
billboard2 = pd.melt(billboard, 
                     id_vars = ['year', 'artist', 'track', 'time', 'date.entered'],
                     var_name = 'week',
                     value_name = 'rating')
billboard2
```

```python
billboard3 = billboard2[['year', 'artist', 'track', 'time']]
billboard3
```

```python
billboard3[billboard3['track'] == 'Loser']
```

```python
billboard3.drop_duplicates()    # 고유한 행 하나를 남기고 나머지 행을 모두 삭제
                                # Series, DataFrame, Index 객체에 적용 가능
```

```python
billboard4 = billboard3.drop_duplicates()
billboard4[billboard4['track'] == 'Loser']
```
