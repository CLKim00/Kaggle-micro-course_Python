# apply() : 내가 만든 함수를 데이터 프레임에 적용

```python
def my_func1(x):
    return x ** 2
```

```python
def my_func2(x, n):
    return x ** n
```

```python
my_func1(4)
```

```python
my_func2(2, 5)
```

```python
import pandas as pd

df = pd.DataFrame({
    'a' : [10, 20, 30],
    'b' : [20, 30, 40]
})

df
```

```python
df['a']
```

```python
df['a'] ** 2
```

```python
df['a'].apply(my_func1)
```

```python
df['a'].apply(my_func2, n=3)
```

```python
def my_print(x):
    print()
```

```python
df.apply(my_print, axis = 0)    # default axis=0 // 행방향 //axis=1 // 열방향
```

```python
df
```

```python
df.apply(my_print, axis = 1)
```

```python
def my_avg(col):
    a = 0
    for i in col:
        a += i
    return a / len(df)
```

```python
df.apply(my_avg, axis = 1)
```
