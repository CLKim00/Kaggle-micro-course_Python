# 1. person, survey 데이터 프레임 합쳐보기


```python
df = person.merge(survey, left_on = 'ident', right_on = 'person')
del df['ident']
df_result = df[['person', 'personal', 'family', 'quant', 'taken', 'reading']]
df_result
```
